```python
def variables (variable):
    category=train_data[variable]
    number= category.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(number.index, number)
    plt.xticks(number.index,number.index.values)
    plt.ylabel("Frequencies")
    plt.xlabel(variable)
    plt.show()
    print(variable, number)
```

```python
category1=["Survived","Sex","Pclass","Embarked","SibSp","Parch"]
```

```python
for i in category1:
    variables(i)
```

```python
category2=["Cabin", "Name","Ticket"]

for i in category2:
    print(train_data[i].value_counts())
    
```

```python
def hist(variable):
    plt.figure(figsize=(9,3))
    plt.hist(train_data[variable])
    plt.xlabel(variable)
    plt.ylabel("Frequencies")
    plt.title(variable)
    plt.show()
    
    
```

```python
category3=["Fare","Age","PassengerId"]
for i in category3:
    hist(i)
```
